# Area Calculation

METHOD 1

#approx area of Punjab

```python
import scipy.misc
import imageio
from PIL import Image
import numpy as np
import random

img=imageio.imread(r"your file location")#eg:c:\foldername\filename
count_pun=0
count_in=0
count=0
while(count<=10000):
    x=random.randint(0,872)
    y=random.randint(0,852)
    z=0
    if(img[x][y][z]==60):
        count_in=count_in+1
        count=count+1
    else:
        if(img[x][y][z]==80):
            count_pun=count_pun+1
            count=count+1
area_place=(count_pun/count_in)*3287263
print(area_place)
```
